**Matplotlib** é uma biblioteca Python para visualização de dados. Ela permite criar gráficos 2D e 3D, incluindo gráficos de linha, barra, dispersão, histogramas e muito mais. Matplotlib é uma ferramenta popular entre cientistas de dados e outros profissionais que precisam apresentar informações visualmente.

```python
import matplotlib
```

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_samples=200,n_features=1,noise=30)
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```
